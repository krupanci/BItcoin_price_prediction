```python
features=bitcoin_df1.drop(columns=['TIMESTAMP','QUOTE_VOLUME','CLOSE'])
features.columns
print(features.shape)
```

```python
X=features.values
y=bitcoin_df1['CLOSE'].values
X.shape,y1.shape
```

```python
from sklearn.model_selection import train_test_split

# Assuming X and y contain your feature set and target variable (closing price)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_train_scaled = scaler_x.fit_transform(X_train)  # Fit & transform training data
X_val_scaled = scaler_x.transform(X_val)  # Transform validation data (use same scaler)

y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1))  
y_val_scaled = scaler_y.transform(y_val.reshape(-1, 1))
```

```python
X_train_scaled = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_val_scaled = X_val_scaled.reshape((X_val_scaled.shape[0], 1, X_val_scaled.shape[1]))
```

```python
X_train_scaled.shape,X_val_scaled.shape
```

```python
import keras_tuner as kt
from tensorflow import keras
from tensorflow.keras import layers

# Function to build LSTM model
def build_lstm_model(hp):
    model = keras.Sequential()
    
    # Tune the number of LSTM units
    model.add(layers.LSTM(units=hp.Int('units', min_value=32, max_value=256, step=32), 
                           return_sequences=False, 
                           input_shape=(1, X_train.shape[1])))
    
    # Tune the dropout rate
    model.add(layers.Dropout(rate=hp.Float('dropout', min_value=0.1, max_value=0.5, step=0.1)))

    # Output layer
    model.add(layers.Dense(1, activation='linear'))

    # Compile model
    model.compile(optimizer=keras.optimizers.Adam(hp.Choice('learning_rate', [1e-3, 1e-4])),
                  loss='mse',
                  metrics=['mae'])
    
    return model

# Use Keras Tuner for hyperparameter search
tuner = kt.RandomSearch(
    build_lstm_model,
    objective='val_loss',
    max_trials=10,  # Number of different hyperparameter combinations to try
    executions_per_trial=2,  # Number of times to train each model
    directory='lstm_tuning',
    project_name='bitcoin_lstm'
)

# Run hyperparameter search
tuner.search(X_train_scaled, y_train_scaled, epochs=20, validation_data=(X_val_scaled, y_val_scaled), verbose=1)

# Get best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best LSTM units: {best_hps.get('units')}, Dropout: {best_hps.get('dropout')}, Learning Rate: {best_hps.get('learning_rate')}")
```
